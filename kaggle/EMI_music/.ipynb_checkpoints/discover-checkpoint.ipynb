```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
users = pd.read_csv('data/users.csv', sep=',')
user_features = ["RESPID","GENDER","AGE","Q1","Q2","Q3","Q4","Q5","Q6","Q7","Q8","Q9","Q10","Q11","Q12","Q13","Q14","Q15","Q16","Q17","Q18","Q19"]
users = users[user_features]
```

```python
users.shape
```

```python
users.info()
```

```python
users_mean_age = users['AGE'].mean()
users_mean_age
```

```python
users['AGE'].fillna(users_mean_age, inplace=True)
```

```python
users['Q16'].fillna(0, inplace=True)
users['Q18'].fillna(0, inplace=True)
users['Q19'].fillna(0, inplace=True)
users.info()
```

```python
users['GENDER'].value_counts()
users
```

```python
plt.hist(users['AGE'], color='green')
plt.show()
```

## 对性别进行编码

```python
from sklearn.feature_extraction import DictVectorizer
vec = DictVectorizer(sparse=False)

gender = users['GENDER']
gender = vec.fit_transform(gender.to_dict())
gender.info()
```
